```python
import pandas as pd
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import shapely as shp
from matplotlib import patheffects

from shapely.geometry import Point
sns.set_style('whitegrid')

fp = r'tanzania-counties\tza_admbnda_adm1_20181019.shp'
map_df = gpd.read_file(fp) 
map_df_copy = gpd.read_file(fp)
map_df.head()

df = pd.read_csv('tanzania_heatmap.csv')
pd.set_option('display.max_columns', None)

#Merging the data
merged = map_df.set_index('ADM1_EN').join(df.set_index('Region'))
merged['Registered Farmers 2022'] = merged['Registered Farmers 2022'].replace(np.nan, 0)
merged.head()

# Create dictionary of state abbreviations
state_abbr = {'Arusha': 'ARU', 'Dar-es-salaam': 'DAR', 'Dodoma': 'DOD', 'Geita': 'GEI', 'Iringa': 'IRI', 
              'Kagera': 'KAG', 'Kaskazini Pemba': 'KPE', 'Kaskazini Unjuga': 'KUN', 'Katavi': 'KAT', 'Kigoma': 'KIG', 
              'Kilimanjaro': 'KIL', 'Kusini Pemba': 'KUP', 'Kusini Unguja': 'KUN', 'Lindi': 'LIN', 'Manyara': 'MAN', 
              'Mara': 'MAR', 'Mbeya': 'MBE', 'Mjini Magharibi': 'MJI', 'Morogoro': 'MOR', 'Mtwara': 'MTW', 'Mwanza': 
              'MWA', 'Njombe': 'NJM', 'Pwani': 'PWA', 'Rukwa': 'RUK', 'Ruvuma': 'RUV', 'Shinyanga': 'SHI', 'Simiyu': 
              'SIM', 'Singida': 'SIN', 'Songwe': 'SON', 'Tabora': 'TAB', 'Tanga': 'TAN'}

# Replace full state names with abbreviations in map_df_copy
map_df_copy['ADM1_EN'] = map_df_copy['ADM1_EN'].replace(state_abbr)

#Create figure and axes for Matplotlib and set the title
fig, ax = plt.subplots(1, figsize=(16, 16))
ax.axis('off')
ax.set_title('Registered Farmers 2022', fontdict={'fontsize': '20', 'fontweight' : '10'})
# Plot the figure
merged.plot(column='Registered Farmers 2022',cmap='YlOrRd', vmax = 16000, linewidth=0.8, ax=ax, edgecolor='0',legend=True,markersize=[39.739192, -104.990337], legend_kwds={'label': "Count"})

# Add county values to the plot
for idx, row in merged.iterrows():
    if row['Registered Farmers 2022'] > 0:
        # Add white drop shadow
        text = plt.annotate(text=str(int(row['Registered Farmers 2022'])),
                            xy=(row['geometry'].centroid.x, row['geometry'].centroid.y - 0.2),
                            horizontalalignment='center',
                            fontsize=8,
                            color='black')
        text.set_path_effects([patheffects.withStroke(linewidth=3, foreground='white')])  # Add white drop shadow

# Add state labels to the plot with a drop shadow effect
for idx, row in map_df_copy.iterrows():
    if row['ADM1_EN'] in state_abbr.values():
        txt = plt.annotate(text=row['ADM1_EN'], xy=row['geometry'].centroid.coords[0], horizontalalignment='center', fontsize=8)
        txt.set_path_effects([patheffects.withStroke(linewidth=3, foreground='white')])
```
